```python
from collections import Counter
import re

# expanded = expanded.filter(expanded["event_type"] == "geopolitical")

# Flatten the "title" column into a single string
all_text = " ".join(expanded["title"].to_list())

# Tokenize the text (words only, ignoring punctuation and case)
words = re.findall(r'\b\w+\b', all_text.lower())

# Count occurrences using Counter
word_counts = Counter(words)

# Convert to a Polars DataFrame for display
word_counts_df = pl.DataFrame(
    {
        "word": list(word_counts.keys()),
        "count": list(word_counts.values())
    }
).sort("count", descending=True)

# Display the result
word_counts_df
```
